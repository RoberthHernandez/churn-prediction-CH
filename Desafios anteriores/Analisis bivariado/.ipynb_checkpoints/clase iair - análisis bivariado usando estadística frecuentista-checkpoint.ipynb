```python
# Configure una función de ayuda para verificar los valores p contra un nivel alfa e imprimir el resultado
def check_p_val(p_val, alpha, type_test=1):
    
    if type_test==1:
            if p_val < alpha:
                print('Tenemos evidencia para rechazar la hipótesis nula')
            else:
                print('No tenemos evidencia para rechazar la hipótesis nula')
    else:
            if abs(p_val) < abs(alpha/2):
                print('Tenemos evidencia para rechazar la hipótesis nula')
            else:
                print('No tenemos evidencia para rechazar la hipótesis nula')
        
# función creada por mi para traer los NA's
def get_na(df):
    qsna=df.shape[0]-df.isnull().sum(axis=0)
    qna=df.isnull().sum(axis=0)
    ppna=round(100*(df.isnull().sum(axis=0)/df.shape[0]),2)
    aux= {'datos sin NAs en q': qsna, 'Na en q': qna ,'Na en %': ppna}
    na=pd.DataFrame(data=aux)
    return na.sort_values(by='Na en %',ascending=False)
```

**Correlación de Pearson**

"El coeficiente de correlación de Pearson (PCC, o r de Pearson) es una medida de correlación lineal ampliamente utilizada. A menudo es el primero que se enseña en muchos cursos elementales de estadísticas. Matemáticamente hablando, se define como “la covarianza entre dos vectores, normalizada por el producto de sus desviaciones estándar”."

Fuente <https://www.freecodecamp.org/news/how-machines-make-predictions-finding-correlations-in-complex-data-dfd9f0d87889/>

```python
data = pd.DataFrame(df.dropna(subset=['price', 'points']))
# points es la variable X
mean_points = data.points.mean()
std_points = data.points.std()
x = data.points
# variable Y va a ser 
mean_price = data.price.mean()
std_price = data.price.std()
y = data.price
N = df.shape[0]
```

```python
#cálculo "a mano" de la covarianza entre points y price
cov_points_price = np.sum((x-mean_points) * (y - mean_price)) / (N-1)
cov_points_price
```

```python
correlacion_points_price = cov_points_price/(std_points*std_price)
correlacion_points_price 
```

**Condiciones para aplicar el coeficiente de correlación de Pearson**

Las condiciones que se deben de cumplir para que el coeficiente de correlación de Pearson sea válido son:

* La relación que se quiere estudiar es de tipo lineal (de lo contrario, el coeficiente de Pearson no la puede detectar).

* Las dos variables deben de ser numéricas.

* Normalidad: ambas variables se tienen que distribuir de forma normal. En la práctica, **se suele considerar válido aun cuando se alejan moderadamente de la normalidad**

* Homocedasticidad: la varianza de  𝑌 debe ser constante a lo largo de la variable  𝑋 Esto se puede contrastar si en un scatterplot los valores de  𝑌 mantienen la misma dispersión en las distintas zonas de la variable  𝑋


**¿Qué es la homocedasticidad?**

La homocedasticidad implica que la varianza se mantiene constante. Puede analizarse de forma gráfica representando las observaciones en un diagrama de dispersión y viendo si mantiene una homogeneidad en su dispersión a lo largo del eje X. Una forma cónica es un claro indicativo de falta de homocedasticidad. Dos test estadísticos utilizados para contrastar la homocedasticidad son: test de Goldfeld-Quandt y el de Breusch-Pagan.

```python
# Correlación lineal entre las dos variables
# ==============================================================================
data = df.dropna(subset=['price', 'points'])
corr_test = pearsonr(x = data['price'], y =  data['points'])
print("Coeficiente de correlación de Pearson: ", corr_test[0])
# El test estadístico exige normalidad entre las variables por lo tanto este test estadístico no es válido
print("P-value: ", corr_test[1])
```

**Correlación de Spearman**

Utiliza la distancia entre las posiciones relativas de los datos una vez que han sido ordenados de menor a mayor dentro del rango de la variable de la muestra. 

Ventajas:

* Se emplea con variables ordinales y/o cuando las variables son continuas pero no satisfacen la condición de normalidad
* Al trabajar con el orden de las observaciones en lugar de su valor real, tiene la característica de ser menos sensible que Pearson a valores extremos




**Condiciones para su aplicación**

* Variables deben ser numéricas continuas y/o ordinales
* El coeficiente de Spearman requiere que la relación entre las variables sea monótona, es decir, que cuando una variable crece la otra también lo hace o cuando una crece la otra decrece (que la tendencia sea constante) 

```python
# Calculate the spearman rank correlation between our data
r_val, p_val = spearmanr(data.price, data.points)

# Check the results of the spearman correlation
print("The correlation coefficient is {:1.4f} with a p-value of {:1.2f}.".format(r_val, p_val))
```

```python
# Plot the data with with expected distribution, under the hypothesis that it is normally distributed. 
#  The blue line is the expected data density, with the same mean and standard deviation, if the data are normal. 
points = df_variety_mean.points.to_numpy()
xs = np.arange(df_variety_mean.points.min(), df_variety_mean.points.max(), 0.01)
fit = stats.norm.pdf(xs, np.mean(points), np.std(points))
plt.plot(xs, fit, label='Normal Dist.', lw=3)
plt.hist(points, 50, density=True, label='Actual Data');
plt.legend();
```
